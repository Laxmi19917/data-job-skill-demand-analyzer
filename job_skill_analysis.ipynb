```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\LAXMI PRASANNA\Downloads\australia-data-job-skill-demand-analyzer\Data\Jobs.csv")
df.head()
```

```python
skill_columns = ["Python", "SQL", "PowerBI", "Tableau", "Excel", "AWS", "MachineLearning"]

for col in skill_columns:
    df[col] = df[col].map({"Yes": 1, "No": 0})

df.head()
```

```python
skill_counts = df[skill_columns].sum().sort_values(ascending=False)
skill_counts
```

```python
import os

os.makedirs("images", exist_ok=True)

plt.figure(figsize=(10,6))
skill_counts.plot(kind="bar")

plt.title("Most Demanded Data Skills in Australia")
plt.xlabel("Skills")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("images/skill_demand.png")   # ✅ FIXED

plt.show()
```

```python
city_counts = df["City"].value_counts()
city_counts
```

```python
plt.figure(figsize=(8,5))
city_counts.plot(kind="bar")
plt.title("Top Hiring Cities for Data Roles")
plt.xlabel("City")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
title_counts = df["Job Title"].value_counts()
title_counts
```

```python
plt.figure(figsize=(10,5))
title_counts.plot(kind="bar")
plt.title("Most Common Data Job Roles")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
